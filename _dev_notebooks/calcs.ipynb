```python
import numpy as np
```

```python
kitty = {"tax": 10000}
user = "test"
users = {"test":{"GLD": 100}}
```

```python
def steal_calc(tax, user_gold, size=1):
    upper_bound = np.min((int(tax) - 9, int(user_gold) + 9))
    lower_bound = np.ceil(upper_bound/2)
    probabilities = np.exp(-np.arange(lower_bound, upper_bound+1)/1000)
    probabilities /= probabilities.sum()
    return np.random.choice(np.arange(lower_bound,upper_bound+1), size=size, p=probabilities)
```

```python
steal_calc(kitty["tax"], users[user]["GLD"])
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
attempts = 100_000
sns.set_theme()
plt.hist(steal_calc(kitty["tax"], users[user]["GLD"], size=attempts))
plt.ylabel("Count")
plt.xlabel("GLD")
plt.title(f"Amount Stolen - {attempts:,} attempts\nTax={kitty['tax']:,}   PlayerGLD={users[user]['GLD']:,}")
```
